## Annotators:

- in simple words these are metadata (data about data), labels or extra information.
- example. 1. i left the room or 2. left of the room , in both sentences left carry different meaning but when we convert into numbers then it may carry same meaning, so for that if we add labels like in first left is verb and Second left is noun, then it may capture this context as well.

<h2>Advantanges</h2>

- these labels can work as feature extraction in ML appraoch
- since we have given labels to the words like POS, NER then this will become structured labels and data, add some patterns in that data for ML approach and that is what ML is looking for, ML is all about learning patterns from the data.

### Types of Annotators
1. NER (Named Entity Recognition)
2. POS (parts of speech ) tagging
3. sentiment analysis (custom annotators) like no of positive or negative words for heuristic approach

#### HOW?
- NER -> using Hidden markov model + viterbi algorithm
    - calculate emission probabilities that is P(word/entity)=P(probability of word give entity) = (no of times that word appeared in document as that entity)/(total number of entities(E)).
    - calculate transition probablity P(E1/E0) probability of entity1 given entity0.
    - final probabilty , P(E/W) = multiply(P(W/E).P(Ei/Ei-1))

    - hidden markov model try all possible combination like location, data, time then there are 8 possible combination for these 3 entities, so it is very expensive
    - to optimize this viterbi algorithm is used , it eliminate all the paths which is giving zero probability.

- POS -> using Hidden markov model + viterbi algorithm
    - calculate emission probabilities that is P(word/pos)=P(probability of word give pos) = (no of times that word appeared in document as that pos)/(total number of entities(E)).
    - calculate transition probablity P(E1/E0) probability of pos1 given pos0.
    - final probabilty , P(E/W) = multiply(P(W/E).P(Ei/Ei-1))

    - hidden markov model try all possible combination like Sub, Verb, Modal then there are 8 possible combination for these 3 entities, so it is very expensive
    - to optimize this viterbi algorithm is used , it eliminate all the paths which is giving zero probability.

## 1. NER 

```python
import spacy

nlp = spacy.load("en_core_web_sm")

text = "Google wes founded by Larry Page and Sergey Brin."
```

```python
doc = nlp(text)
print(doc.text)
tokens = []
for token in doc:
    tokens.append(token.text)
NER = []
for token in doc.ents:
    NER.append((token.text, token.label_))
    
print(tokens)
print(NER)
```

```python
for entity in NER:
    print(f"{entity[0]} : {entity[1]} : {spacy.explain(entity[1])}")
```

## 2. POS Tagging

```python
import nltk
from nltk import pos_tag, word_tokenize

# nltk.download('averaged_perceptron_tagger_eng')
```

```python
text = "Will Will search my blogs"
tokens = word_tokenize(text)
tagged = pos_tag(tokens)
for tag in tagged:
    print(tag)
```

```python
text = "left of the room"
tokens = word_tokenize(text)
tagged = pos_tag(tokens)
for tag in tagged:
    print(tag)
```

**Excellent result**

<strong>Using SpaCy Library:</strong>

```python
import spacy

nlp = spacy.load('en_core_web_sm')
```

```python
doc = nlp(text)
print(doc)
for word in doc:
    print(word, "-->", word.pos_, word.tag_, spacy.explain(word.tag_))
```

```python
for word in nlp("left of the room."):
    print(word, "-->", word.tag_, spacy.explain(word.tag_))
```
