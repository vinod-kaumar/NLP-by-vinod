````python
import numpy as np
import pandas as pd
import torch
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
````

````python
df = pd.read_csv('https://raw.githubusercontent.com/gscdit/Breast-Cancer-Detection/refs/heads/master/data.csv')
df.head()
````

````python
df.drop(columns=['id', 'Unnamed: 32'], inplace= True)
````

````python
X_train, X_test, y_train, y_test = train_test_split(df.iloc[:, 1:], df.iloc[:, 0], test_size=0.2)
````

````python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
````

````python
encoder = LabelEncoder()
y_train = encoder.fit_transform(y_train)
y_test = encoder.transform(y_test)
````

# Numpy arras to pytorch tensors

````python
X_train_tensor = torch.from_numpy(X_train)
X_test_tensor = torch.from_numpy(X_test)
y_train_tensor = torch.from_numpy(y_train)
y_test_tensor = torch.from_numpy(y_test)
````

# ANN model

````python
class ANNModel():
  def __init__(self, X):
    self.W = torch.rand(X.shape[1], 1, dtype = torch.float64, requires_grad = True)
    self.b = torch.rand(1, dtype = torch.float64, requires_grad = True)


  def forward(self, X_train):
    z = torch.matmul(X_train, self.W) + self.b
    return torch.sigmoid(z)

  def loss_function(self, y_pred, y):
    # clamp predictions to avoid log(0)
    epsilon = 1e-7
    y_pred = torch.clamp(y_pred, epsilon, 1-epsilon)

    #calculate loss
    return -(y*torch.log(y_pred) + (1-y)*torch.log(1-y_pred)).mean()


````

# parameters

````python
learning_rate = 0.1
epochs = 25
````

# training pipeline

````python
# create model
model = ANNModel(X_train_tensor)
print(model.W.shape)
print(model.b)
````

````python
# forward pass

# define loop

for epoch in range(epochs):
  # forward pass
  y_pred = model.forward(X_train_tensor)
  # print(y_pred.shape)
  # print(y_pred)

  # loss calculation
  loss = model.loss_function(y_pred, y_train_tensor)


  # backward pass
  loss.backward()

  # parameters updation
  with torch.no_grad(): # it will not track grad here
    model.W -= learning_rate * model.W.grad
    model.b -= learning_rate * model.b.grad

  # zero gradients
  model.W.grad.zero_()
  model.b.grad.zero_()

  #print loss in each epoch
  print(f"Epoch: {epoch+1}, Loss: {loss}")
````

# Evaluatoin

````python
# mdoel evaluation
with torch.no_grad():
  y_pred = model.forward(X_test_tensor)

  # print(y_pred)

  y_pred = (y_pred > 0.5).float()

  print(y_pred[:5])
````

````python
from sklearn.metrics import accuracy_score
print("accuracy is: ", accuracy_score(y_test_tensor, y_pred))
````

- Automatically manages learnable parameters (weights and biases).
- Integrates seamlessly with Autograd for gradient computation.
- Reduces the need to manually implement neural network components.
- Acts as the foundation for creating custom neural networks.

# creating new architecture
- input features = 5
- 1 hidden layer with 3 nodes ---> relu
- 1 output layer with 1 node --> sigmoid

````python
# create model
import torch
import torch.nn as nn

class Model(nn.Module):

  def __init__(self, num_features):
    super().__init__() # child invokes constructor of nn module

    self.linear1 = nn.Linear(num_features, 3)
    self.relu= nn.ReLU()
    self.linear2 = nn.Linear(3, 1)
    self.sigmoid = nn.Sigmoid()

  def forward(self, features):
    out = self.linear1(features)
    out = self.relu(out)
    out = self.linear2(out)
    out = self.sigmoid(out)

    return out
````

# `torch.optim` Module

## What is `torch.optim`?

- `torch.optim` is a PyTorch module that provides various optimization algorithms used to update model parameters during training.
- It automates the process of updating weights and biases using computed gradients.
- Common optimizers include:
  - SGD (Stochastic Gradient Descent)
  - Adam
  - RMSProp
  - Adagrad
  - AdamW

### Benefits

- Efficient parameter updates
- Learning rate management
- Weight decay (regularization)
- Momentum support
- Advanced optimization techniques

---

## `model.parameters()`

- `model.parameters()` returns an iterator over all **trainable parameters** in a model.
- These parameters are instances of `torch.nn.Parameter`.
- The optimizer uses these parameters to compute gradients and update them during training.

### Includes

#### Weights
- Weight matrices of layers such as:
  - `nn.Linear`
  - `nn.Conv2d`
  - `nn.LSTM`
  - etc.

#### Biases
- Bias terms of layers (if present).

### Example

```python
for param in model.parameters():
    print(param.shape)
```

---

## Using an Optimizer

```python
import torch.optim as optim

optimizer = optim.SGD(
    model.parameters(),
    lr=0.01

```

### Training Steps

```python
optimizer.zero_grad()

loss.backward()

optimizer.step()
```

### What Happens?

1. `zero_grad()` → Clears old gradients.
2. `backward()` → Computes gradients using Autograd.
3. `step()` → Updates model parameters using the optimizer.
