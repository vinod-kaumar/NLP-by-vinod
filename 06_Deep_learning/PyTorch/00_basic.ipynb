# PyTorch

## What is PyTorch?

- **Open-source Deep Learning Library**
- Developed by **Meta AI (Facebook AI Research)**
- Combines:
  - **Python** (easy syntax)
  - **Torch** (high-performance tensor computations)
- Designed for:
  - Deep Learning
  - Neural Networks
  - GPU Computing

---

# Evolution of PyTorch

## PyTorch 0.1 (2017)

### Features
- Dynamic Computation Graph
- Integration with NumPy & SciPy

### Impact
- Popular among researchers
- Pythonic and flexible

---

## PyTorch 1.0 (2018)

### Features
- Research → Production bridge
- TorchScript introduced
- Caffe2 integration

### Impact
- Easier deployment of models

---

## PyTorch 1.x

### Features
- Distributed Training
- ONNX Support
- Quantization
- Domain Libraries:
  - torchvision
  - torchtext
  - torchaudio

### Impact
- Industry adoption increased
- Hugging Face & PyTorch Lightning ecosystem growth

---

## PyTorch 2.0

### Features
- Faster execution
- Better deployment support
- Optimized for modern hardware

### Impact
- Improved scalability
- Production-ready performance

# PyTorch vs TensorFlow

## PyTorch Advantages

- More Pythonic
- Easier to learn
- Dynamic Graph Execution
- Better for Research
- High Customization
- Strong Python Integration

## TensorFlow Advantages

- Better Production Ecosystem
- TensorFlow Serving
- TensorFlow Lite
- More Mature Deployment Tools
- Better Mobile Support

### Rule of Thumb

- **Research → PyTorch**
- **Large-scale Production → TensorFlow**
- **Today both are widely used**

# PyTorch Fundamentals

## Three Pillars

### 1. Tensors
- Data storage
- Multi-dimensional arrays

### 2. Computation Graph
- Tracks operations
- Enables backpropagation

### 3. Autograd
- Automatic gradient calculation
- Core of neural network training
