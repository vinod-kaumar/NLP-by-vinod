```python
import pandas as pd
import numpy as np
import gensim
import os

import warnings
warnings.filterwarnings("ignore")
```

```python
from nltk import sent_tokenize
from gensim.utils import simple_preprocess
```

```python
story = []
for filename in os.listdir('data'):
    
    f = open(os.path.join('data', filename), 'r')
    corpus = f.read()
    raw_sentences = sent_tokenize(corpus)
    for sentence in raw_sentences:
        story.append(simple_preprocess(sentence))
# simple preprocessing function 
#converts to lowercase
# removes punctuation
# tokenizes into words
```

```python
story[:5]
```

```python
model = gensim.models.Word2Vec(
    window = 10,
    min_count = 2
)
# bydefault vector_size = 100
# sg = 0 --> CBOW default
# sg = 1 --> skip-gram
```

```python
model.build_vocab(story)
# here model reads all the words of story and check it's frequency as well
# if frequency < 2 --> remove it 
# it creates vocabulary dictionary
# initializes Neural Network structures 
```

```python
model.corpus_count
```

```python
model.epochs
```

```python
model.train(story, total_examples = model.corpus_count, epochs = 10)
```
