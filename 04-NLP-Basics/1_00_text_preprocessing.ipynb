```python
# convert numbers with corresponding word
def numbers_to_word(text):
    result = []
    for word in text.split():
        if word.isdigit():
            result.append(num2words(word))
        else:
            result.append(word)
    return " ".join(result)
```

# 6 Remove Punctuations:

```python
import string
exclude = string.punctuation
print("Punctuations are:",exclude)
```

```python
def remove_punctuation(text):
    for char in exclude:
        text = text.replace(char, "")
    return text
```

```python
print(remove_punctuation("hello i am vinod. Learning NLP but why?"))
```

```python
import time
start = time.time()
df['movie_name_1'] = df.movie_name.apply(remove_punctuation)
end = time.time()
time1 = end-start
print("time taken in algorith:", time1,"s")
```

```python
df.head()
```

```python
# using standard library to remove punctuations
def remove_punctuation_fast(text):
    return text.translate(str.maketrans("", "", exclude))
```

```python
print(remove_punctuation_fast("hello i am vinod. Learning NLP but why?"))
```

```python
import time
start = time.time()
df['movie_name'] = df.movie_name.apply(remove_punctuation_fast)
end = time.time()
time2 = end-start
print("time taken in algorith:", time1,"s")
```

```python
time2 < time1
```

```python
# in huge dataset this time difference (time2 - time1) will be significant
# so use this second method
```

# 7 Chat word treatment

- used on social media generated dataset

- when u chat on social media :-

- u : you
- FYI : for your information
- GN : good night
- GM : good morning

```python
map = {
    "U":"You",
    "FYI":"For Your Information",
    "GM":"Good Morning",
    "ASAP":"As Soon As Possible"
}
```

```python
def chat_conversation(text):
    for key, value in map.items():
        text = text.replace(key, value)
    return text
```

```python
text = "U should check this out ASAP! FYI, it's really good. GM!"
print(chat_conversation(text))
```

# 9 Handling Emojis😂

- used when we handle social media dataset

```python
# two ways of handling emojis in text:
# 1. Remove emojis using regex
# 2. replace emojis with its meaning using emoji library
```

```python
def remove_emoji(text):
    emoji_pattern = re.compile("["
                           u"\U0001F600-\U0001F64F"  # emoticons
                           u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                           u"\U0001F680-\U0001F6FF"  # transport & map symbols
                           u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                           u"\U00002702-\U000027B0"
                           u"\U000024C2-\U0001F251"
                           "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', text)
```

```python
remove_emoji("Hello 😊! This is a test. 🚀")
```

```python
from nltk.corpus import stopwords
print(stopwords.words("english"))
```

```python
print(stopwords.words("french"))
```

```python
# stopwords.words("hindi") 
# hindi stop words are not available
```

```python
def remove_stopwords(text):
    filtered_words = [word for word in text.split() if word.lower() not in stopwords.words("english")]
    return " ".join(filtered_words)
```

```python
remove_stopwords("hindi stop words ARE not available")
```

```python
import time
start = time.time()
df['movie_name_1'] = df.movie_name.apply(remove_stopwords)
end = time.time()
time1 = end - start
```

```python
# faster code using sets
stop_words = set(stopwords.words('english'))
def remove_stopwords_fast(text):
    filtered_words = [word for word in text.split() if word.lower() not in stop_words]
    return " ".join(filtered_words)
```

```python
remove_stopwords_fast("hindi stop words ARE not available")
```

```python
import time
start = time.time()
df['movie_name'] = df.movie_name.apply(remove_stopwords_fast)
end = time.time()
time2 = end - start
```

```python
print(f"second algorithm is {time1 - time2}s faster than first one")
```

# Advance Text-Preprocessing

## 1. POS Tagging

- Parts of speech tagging
- it is a preprocessing step in text preprocessing
- it is a task of labelling each word in a sentence with it's appropriate parts of speech
- POS is a category of words that have similar grammatical properties
- ex.. Will will search blogs. here will --> Modal, will --> Noun, search --> Verb and blogs --> Noun
- <p>In  <span style = "background-color:red; color:white">I left<span> the room and <span style = "background-color:red; color:white">left<span> of the room<p>
- both left carry diffferent meaning

<strong>Applications of POS tagging</strong>
1. Named entity recognition used in retrieval system.
2. Question answering system
3. Word sense disambiguation --> i left the room and left of the room , here left has different context in both the sentences
4. chatbots

<strong>We do Part-of-Speech (POS) tagging because it helps a computer understand how words function in a sentence—not just what they mean individually, but how they relate to each other.</strong>

<strong>POS Tagging using nltk:</strong>

```python
import nltk
from nltk import pos_tag

# nltk.download('averaged_perceptron_tagger_eng')
```

```python
text = "Will Will search my blogs"
tokens = word_tokenize(text)
tagged = pos_tag(tokens)
for tag in tagged:
    print(tag)
```

```python
text = "left of the room"
tokens = word_tokenize(text)
tagged = pos_tag(tokens)
for tag in tagged:
    print(tag)
```

## 2. NER

```python
import spacy

nlp = spacy.load("en_core_web_sm")

text = "Google wes founded by Larry Page and Sergey Brin."
```

```python
doc = nlp(text)
print(doc.text)
tokens = []
for token in doc:
    tokens.append(token.text)
NER = []
for token in doc.ents:
    NER.append((token.text, token.label_))
    
print(tokens)
print(NER)
```

```python
for entity in NER:
    print(f"{entity[0]} : {entity[1]} : {spacy.explain(entity[1])}")
```

## 4. Coreference Resolution

- Coreference resolution means finding which words or phrases refer to the same real-world entity in a text.

**Ex.**
- “Rahul went to the store. He bought milk.”

- Here:
    * Rahul
    * He
- refer to the same person.
- The NLP system must understand that “He” = “Rahul”.
- Humans naturally understand references, but computers do not.
