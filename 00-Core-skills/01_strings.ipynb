```python
# slicing and indexing strings
name = "vinod"
name[0]
```

```python
print(name[-1])
print(name[1:4]) # slicing from index 1 to 3 (4 is exclusive)
print(name[:3]) # slicing from start to index 2 (3 is exclusive)
print(name[2:]) # slicing from index 2 to end
print(name[:-2]) # slicing from start to index -3 (negative indexing)
```

```python
# reversing a string
print(name[::-1])
```

```python
# strings are immutable
# name[0] = "v" # this will raise an error because we cannot change a character in a string
```

```python
try:
    name[0] = "v"
except TypeError as e:
    print(e)
```

```python
# no of characters in a string = length of the string including spaces
print(len(name))
```

```python
# use case - 1 : cleaning text data 
# short form to long form
text = "I'm a student. I'm learning NLP. Gnt, see you tomorrow."
cleaned_text = text.replace("I'm", "I am").replace("Gnt", "Good night")
```

```python
# string searching 
search_word = "student"
if search_word in cleaned_text:
    print(f"{search_word} found in text")
else:
    print(f"{search_word} not found in text")
```

```python
product_codes = ["P002", "P003", "P001"]
product_prices = {"P001": 10.0, "P002": 20.0, "P003": 30.0}
scanned_code = input("enter product code: ")
# scanned_code = "P002"

if scanned_code in product_codes:
    print("Product code is valid")
    print(f"Price: ${product_prices[scanned_code]:.2f}")
else:
    print("Product code is invalid")
```
