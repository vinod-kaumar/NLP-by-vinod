### Train test spliting

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 42)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

### Scaling

### standard scaler from scratch

```python
class StandardScaler1:
    def fit(self, X):
        self.mean = np.mean(X, axis = 0)
        self.std  = np.std(X, axis = 0)
        
    def transform(self, X):
        return (X - self.mean)/self.std
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_train_scaled = scaler.fit_transform(X_train[num_cols_new])
X_test_scaled = scaler.transform(X_test[num_cols_new])
```

```python
X_train_scaled = pd.concat([pd.DataFrame(X_train_scaled, columns = num_cols_new), pd.DataFrame(X_train.drop(num_cols_new, axis = 1).values, columns = X_train.drop(num_cols_new, axis = 1).columns)], axis = 1   )
```

```python
X_test_scaled = pd.concat([pd.DataFrame(X_test_scaled, columns = num_cols_new), pd.DataFrame(X_test.drop(num_cols_new, axis = 1).values, columns = X_test.drop(num_cols_new, axis = 1).columns)], axis = 1   )
```

```python
X_train_scaled.shape, X_test_scaled.shape
```

```python
scaler_1 = StandardScaler1()
scaler_1.fit(X_train[num_cols_new])
X_train_scaled_1 = scaler_1.transform(X_train[num_cols_new])
X_test_scaled_1 = scaler_1.transform(X_test[num_cols_new])
```

```python
scaler_1.mean, scaler_1.std
```

```python
X_train_scaled_1 = pd.concat([X_train_scaled_1, X_train.drop(columns = num_cols_new)], axis = 1   )
X_test_scaled_1 = pd.concat([X_test_scaled_1, X_test.drop(columns = num_cols_new)], axis = 1   )
```

```python
X_train_scaled.shape, X_train_scaled_1.shape
```

**Here data is cleaned and prepared for model training, u can use both scaler_1 or scaler**
