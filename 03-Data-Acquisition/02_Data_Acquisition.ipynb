```python
# here it is providing genre_ids instead of genre names, we need to convert them to genre names using the genre ids and genre names mapping provided by the api.
```

```python
genre_url = f"https://api.themoviedb.org/3/genre/movie/list?api_key={API_KEY}&language=en-US"
genre_response = requests.get(genre_url)
genre_response.json()
```

```python
genre_dict = {}
for genre in genre_response.json()['genres']:
    genre_dict[genre["id"]] = genre["name"]
```

```python
df = pd.DataFrame(response.json()['results'])[['title', 'overview', 'genre_ids']]
```

```python
df.head()
```

```python
# now we are fetching data of all the 428 pages
df = pd.DataFrame()
for i in range(1, 429):
    url = f"https://api.themoviedb.org/3/movie/top_rated?api_key={API_KEY}&language=en-US&page={i}"
    response = requests.get(url)
    temp_df = pd.DataFrame(response.json()['results'])[['title', 'overview', 'genre_ids']]
    df = pd.concat([df, temp_df], ignore_index = True)
```

```python
genre_dict
```

```python
genres = []
for ids in df.genre_ids:
    genre_names = [genre_dict[id] for id in ids]
    genres.append(genre_names)

df["genre"] = genres

df.rename(
    columns = {
        "overview": "description",
        "title": "movie_name"
    }, inplace = True
)
```

```python
df.drop("genre_ids", axis = 1, inplace = True)
```
