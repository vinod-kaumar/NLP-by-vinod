## **Tensors**

- Tensor is a specialized multi-dimensional array designed for mathematical and computational efficiency.

---

## **Real-World Examples**

### **1. Scalars (0-Dimensional Tensors)**
- A single number.
- Example:
  - Loss value after the forward pass.
  - `5.0`, `-4.5`

---

### **2. Vectors (1-Dimensional Tensors)**
- Represents a sequence or collection of values.
- Example:
  - Feature vector in NLP (word embeddings).
  - `[0.12, -0.34, 0.33]`

---

### **3. Matrices (2-Dimensional Tensors)**
- Represents tabular or grid-like data.
- Example:
  - Grayscale image represented as a 2D tensor.

---

### **4. 3D Tensors**
- Adds a third dimension, often used for stacking data.
- Example:
  - RGB (Colored) images.

---

### **5. 4D Tensors**
- Adds the batch size as an additional dimension to 3D data.
- Example:
  - Batch of RGB images.
  - Shape: `[batch_size, width, height, channels]`

---

### **6. 5D Tensors**
- Adds a time dimension for data that changes over time.
- Example:
  - Video data (sequence of RGB frames).
  - Shape: `[batch_size, frames, width, height, channels]`
  - Example Shape: `[10, 16, 64, 64, 3]`
    - `10` → Video clips (batch size)
    - `16` → Frames per video
    - `64 × 64` → Frame resolution
    - `3` → RGB channels

## **Why are Tensors Useful?**

### **1. Mathematical Computations**
- Tensors enable efficient mathematical computations such as:
  - Addition
  - Multiplication
  - Dot Product
- These operations are essential for neural network computations.

---

### **2. Representation of Real-World Data**
- Real-world data such as images, audio, and text can be represented as tensors.

#### Examples:
- **Images:** Represented as 3D tensors `(Width × Height × Channels)`
- **Text:** Tokenized and represented as 2D or 3D tensors

---

### **3. Efficient Computations**
- Tensors are optimized for hardware acceleration.
- They can be processed efficiently on:
  - GPUs (Graphics Processing Units)
  - TPUs (Tensor Processing Units)
- This makes training deep learning models significantly faster.

---

## **Where are Tensors Used in Deep Learning?**

### **1. Data Storage**
- Training data such as:
  - Images
  - Text
  - Audio
- is stored in the form of tensors.

---

### **2. Weights and Biases**
- The learnable parameters of neural networks are stored as tensors.
- Examples:
  - Weight matrices
  - Bias vectors

---

### **3. Matrix Operations**
- Neural networks heavily rely on tensor operations such as:
  - Matrix Multiplication
  - Dot Products
  - Broadcasting
- All these operations are performed using tensors.

---

### **4. Training Process**

#### **Forward Pass**
- Tensors flow through the network layer by layer.
- Computations are performed to generate predictions.

#### **Backward Pass**
- Gradients are computed during backpropagation.
- These gradients are also represented as tensors and are used to update model parameters.

# 2. element wise operations

```python
A = torch.rand(3, 3)
B = torch.rand(3, 3)

print("Tensor A:\n", A)
print("\nTensor B:\n", B)


print("\nElement-wise Addition (A + B):")
print(A + B)

print("\nElement-wise Subtraction (A - B):")
print(A - B)

print("\nElement-wise Multiplication (A * B):")
print(A * B)

print("\nElement-wise Division (A / B):")
print(A / B)

print("\nElement-wise Power (A ** 2):")
print(A ** 2)


print("\nA > B:")
print(A > B)

print("\nA == B:")
print(A == B)


print("\nSquare Root of A:")
print(torch.sqrt(A))

print("\nExponential of A:")
print(torch.exp(A))

print("\nNatural Log of A:")
print(torch.log(A))
```

## Inplace operations

```python
import torch

A = torch.tensor([1, 2, 3], dtype=torch.float32)

print("Original:", A)

# In-place addition
A.add_(5)
print("\nAfter add_:", A)

# In-place subtraction
A.sub_(2)
print("\nAfter sub_:", A)

# In-place multiplication
A.mul_(3)
print("\nAfter mul_:", A)

# In-place division
A.div_(2)
print("\nAfter div_:", A)

# In-place power
A.pow_(2)
print("\nAfter pow_:", A)
```

## Copying a tensor

```python
A = torch.tensor([1, 2, 3])

B = A

B[0] = 100

print("A:", A)
print("B:", B)

# here A is changed , beccause of same id
print(id(A))
print(id(B))
```

```python

# Copy tensor
B = A.clone()

B[0] = 100

print("A:", A)
print("B:", B)
```

## comparision between cpu and gpu tensors

```python
import torch
import time

size = 5000

# CPU benchmark
A_cpu = torch.rand(size, size)
B_cpu = torch.rand(size, size)

start = time.time()
C_cpu = A_cpu @ B_cpu
cpu_time = time.time() - start

print(f"CPU Time: {cpu_time:.4f} seconds")

# GPU benchmark
if torch.cuda.is_available():

    A_gpu = A_cpu.to("cuda")
    B_gpu = B_cpu.to("cuda")

    torch.cuda.synchronize()
    start = time.time()

    C_gpu = A_gpu @ B_gpu

    torch.cuda.synchronize()
    gpu_time = time.time() - start

    print(f"GPU Time: {gpu_time:.4f} seconds")
    print(f"Speedup: {cpu_time / gpu_time:.2f}x")

else:
    print("CUDA is not available")
```
