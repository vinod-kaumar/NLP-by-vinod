# Feature Extraction (Text representation)

### What?
- Converts text into numerical form so machines can process it
- Extracts important information and patterns from text
- Also known as text vectorization
- Helps capture the meaning and context of words/sentences
- Required because computers work with numbers, not raw text
- Used as a preprocessing step in many Natural Language Processing tasks like classification, sentiment analysis, and translation

### Why?
- We extract features from text and convert them into numerical form
- Computers can process only numbers, not raw text
- The quality of features strongly affects model performance
- Bad features can lead to poor results even with a powerful model
- Good features can help a simple or average model perform very well
- In many cases, better feature extraction is more important than using a more complex model

### Core Idea
- The main goal is to convert text into numerical form while preserving its semantic meaning
- Similar words or sentences should have similar numerical representations
- The numerical representation should help machines understand - context and relationships between words
- This forms the foundation of many Natural Language Processing tasks like sentiment analysis, translation, and text classification

### Techniques

#### 1. One-Hot Encoding
- Represents each word as a unique binary vector

#### 2. Bag of Words (BoW)
- Converts text into vectors based on word frequency

#### 3. n-grams
- Captures sequences of words to preserve some context

#### 4. TF-IDF (Term Frequency–Inverse Document Frequency)
- Gives higher importance to meaningful and less frequent words

#### 5. Custom Features using Domain Knowledge
- Manually created features based on specific problem understanding

#### 6. Word2Vec / Word Embeddings
- Deep learning-based representations where similar words have similar vector meanings


### Common Terms

- **Corpus**: Collection of all text data or all words present in the dataset (C)

- **Vocabulary**: Set of unique words present in the corpus (V)

- **Document**: An individual sentence, paragraph, or text sample from the dataset (D)

- **Word**: Basic unit of text; every document is made up of words
