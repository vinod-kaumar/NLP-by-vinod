# POS
- Parts of speech tagging

- it is a preprocessing step in text preprocessing
- it is a task of labelling each word in a sentence with it's appropriate parts of speech
- POS is a category of words that have similar grammatical properties
- ex.. Will will search blogs. here will --> Modal, will --> Noun, search --> Verb and blogs --> Noun

### Applications of POS tagging

1. Named entity recognition used in retrieval system.
2. Question answering system
3. Word sense disambiguation --> i left the room and left of the room , here left has different context in both the sentences
4. chatbots

### Code 

```python
import numpy as np
import pandas as pd

import warnings
warnings.filterwarnings("ignore")
```

```python
import spacy
nlp = spacy.load("en_core_web_sm")
```

```python
doc = nlp("Apple is looking at buying U.K. startup for $1 billion")
```

```python
doc.text
```

```python
doc[0]
```

```python
doc[0].pos_
```

```python
spacy.explain("PROPN")
```

```python
for word in doc:
    print(word.text, "----->", word.pos_, word.tag_, spacy.explain(word.tag_))
```

```python
doc2 = nlp("I left the room")
for word in doc2:
    print(word.text, "----->", word.pos_, word.tag_, spacy.explain(word.tag_))
```

```python
doc3 = nlp("to the left of the room")
for word in doc3:
    print(word.text, "----->", word.pos_, word.tag_, spacy.explain(word.tag_))
```

```python
doc5 = nlp("will will search blogs")
```
