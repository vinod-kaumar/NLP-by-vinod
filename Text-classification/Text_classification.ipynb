## Text Preprocessing

### 1. Lowercasing

```python
df['review'] = df['review'].str.lower()
```

```python
df.head()
```

### 2. Removing HTML tags

```python
df.review[0]
```

```python
import re
def remove_tags(text):
  pattern = r'<.*?>'
  return re.sub(pattern, "", text);
```

```python
df['review'] = df['review'].apply(remove_tags)
```

```python
df["review"][0]
```

### 3. Removing URLs

```python
import re

def remove_urls(text):
    pattern = r'https?://\S+|www\.\S+'
    text = re.sub(pattern, '', text)
    print(re.findall(pattern, text))
    return re.sub(r'\s+', ' ', text).strip()
```

```python
df['review'] = df['review'].apply(remove_urls)
```

### 4. Remove Punctuations

```python
df.head()
```

```python
import string
exclude = string.punctuation
```

```python
def remove_punctuations(text):
  return text.translate(str.maketrans('', '', exclude))
```

```python
import time
start = time.time()
df['review'] = df['review'].apply(remove_punctuations)
end = time.time()
print(end - start)
```

### 7. Removing Stop Words

```python
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))

def remove_stopwords(text):
    filtered_words = [
        word for word in text.split()
        if word.lower() not in stop_words
    ]
    return ' '.join(filtered_words)
```

```python
df['review'] = df['review'].apply(remove_stopwords)
```

### 8. Demojize

```python
def text_demojize(text):
  return emoji.demojize(text)
```

```python
df['review'] = df['review'].apply(text_demojize)
```

### 9. Stemming

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem_words(text):
  return " ".join(ps.stem(word) for word in text.split())
```

```python
df['review'] = df['review'].apply(stem_words)
```

# Modelling - ML approach

## 1. with BOW

```python
X = df.drop('sentiment', axis = 1)
y = df['sentiment']
```

```python
X.shape, y.shape
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()

y = encoder.fit_transform(y)
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test , y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 2)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# applying BOW
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000) # if we take frequent 5k words accuracy is good if we increase it decreases
```

```python
X_train_bow = cv.fit_transform(X_train['review']).toarray()
```

```python
X_train_bow.shape
```

```python
X_test_bow = cv.transform(X_test['review']).toarray()
```

```python
X_test_bow.shape
```

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
```

```python
gnb.fit(X_train_bow, y_train)
```

```python
y_pred = gnb.predict(X_test_bow)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
accuracy_score(y_test, y_pred)
```

```python
confusion_matrix(y_test, y_pred)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()

rf.fit(X_train_bow, y_train)
```

```python
y_pred = rf.predict(X_test_bow)
accuracy_score(y_test, y_pred)
```

```python
confusion_matrix(y_test, y_pred)
```

## 2. using ngrams

```python
# applying ngrams
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(ngram_range = (2,2), max_features = 5000)

X_train_bow = cv.fit_transform(X_train['review']).toarray()
X_test_bow = cv.transform(X_test['review']).toarray()

from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()

gnb.fit(X_train_bow, y_train)

y_pred = gnb.predict(X_test_bow)

print(accuracy_score(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```

```python
# applying ngrams
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(ngram_range = (1,2), max_features = 5000)

X_train_bow = cv.fit_transform(X_train['review']).toarray()
X_test_bow = cv.transform(X_test['review']).toarray()

from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()

gnb.fit(X_train_bow, y_train)

y_pred = gnb.predict(X_test_bow)

print(accuracy_score(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```
