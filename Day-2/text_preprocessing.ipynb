## Regular expressions, Regex

```python
import re

# Match the word "python" in the text
text = "I love python programming!"
match = re.search(r'python', text)
print(match.group())
```

```python
# The re.search(r'python', text) function searches for the word "python" in the given text.
# The regular expression r'python' is looking for the exact match of the string "python".
# If a match is found, re.search() returns a match object.
# match.group() retrieves the matched string, in this case, "python"
```

- Identifying specific words or phrases in text, like extracting keywords or entities.

### Metacharacters

Metacharacters are special characters that define the behavior of the regex engine.

Common Metacharacters:

. - Matches any character except newline.

^ - Matches the start of a string.

$ - Matches the end of a string.

[] - Matches a character set.

| - Acts as an OR between patterns.

```python
def match_pattern(pattern, text):
    match = re.search(pattern, text)
    if match:
        return match.group()
    else:
        return "No match found"
```

```python
print(match_pattern(r'hello.w', "helloworld"))
print(match_pattern(r'hello.w', "hello world!"))
```

```python
text = "The price is 100"
# Caret (^): Matches the start of the string.
print(match_pattern(r"^The", text))
print(match_pattern(r"^100", text))
# Dollar ($): Matches the end of the string.
print(match_pattern(r"The$", text))
print(match_pattern(r"100$", text))
# Asterisk (*): Matches 0 or more repetitions of the preceding pattern.
print(match_pattern(r"pri*", text))
print(match_pattern(r"llo*", "helloooo world"))
# Plus (+): Matches 1 or more repetitions of the preceding pattern.
print(match_pattern(r"pri+", text))
print(match_pattern(r"lo+", "helloooo world"))
print(match_pattern(r"llo+", "helloooo world"))
print(match_pattern(r"ll+", "helloooo world"))
```

- Matching patterns at the start or end of sentences for sentence boundary detection

###  Character Classes
Character classes allow you to match one of a set of characters.

[abc] - Matches 'a', 'b', or 'c'.

[^abc] - Matches any character except 'a', 'b', or 'c'.

```python
print((re.findall(r'[aeiou]', "hello worldu")))
print(re.findall(r'[^aeiou]', "hello worldu"))
```

- Extracting parts of text like named entities (e.g., extracting names from sentences)

### Quantifiers
Quantifiers specify how many times a pattern should appear.

-> * - Matches 0 or more of the preceding pattern.

-> + Matches 1 or more of the preceding pattern.

? - Matches 0 or 1 of the preceding pattern.

{n} - Matches exactly n occurrences of the preceding pattern.

{n,} - Matches n or more occurrences of the preceding pattern.

{n,m} - Matches between n and m occurrences of the preceding pattern.

```python
match = re.findall(r'\d{1,4}', "1234 and 56")
print(match)
```

```python
match = re.findall(r'\d{1,4}', "12345 and 56 55555")
print(match)
```

# 4. Remove punctuations

```python
import string
exclude = string.punctuation
```

```python
exclude
```

```python
def remove_punctuation(text):
    for char in exclude:
        text = text.replace(char, "")
    return text
```

```python
import time
start = time.time()
print(remove_punctuation("Hello, world! This is a test."))
end = time.time()
print(f"Time taken: {end - start}")
time1 = end - start
```

```python
# above method is slow for large texts. We can use standard library to remove punctuation
def remove_punctuation_fast(text):
    return text.translate(str.maketrans('', '', exclude))
```

```python
start = time.time()
print(remove_punctuation_fast("Hello, world! This is a test."))
end = time.time()
print(f"Time taken: {end - start}")
time2 = end - start
```

```python
import time
start = time.time()
df['sentence_punctuation'] = df['sentence'].apply(remove_punctuation)
end = time.time()
print(f"Time taken: {end - start}")
time1 = end - start
```

```python
import time
start = time.time()
df['sentence_punctuation_fast'] = df['sentence'].apply(remove_punctuation_fast)
end = time.time()
print(f"Time taken: {end - start}")
time2 = end - start
```

```python
time1/time2
# time1 is much higher than time2, which shows that the second method is much faster than the first method for removing punctuation from text.
```

# 5. Chat word treatment

- when u chat on social media :-

- u : you
- FYI : for your information
- GN : good night
- GM : good morning

```python
map = {
    "U":"You",
    "FYI":"For Your Information",
    "GM":"Good Morning",
    "ASAP":"As Soon As Possible"
}
```

```python
text = "U should check this out ASAP! FYI, it's really good. GM!"
```

```python
def chat_conversation(text):
    for key, value in map.items():
        text = text.replace(key, value)
    return text
```

```python
print(chat_conversation(text))
```

```python
# faster code using sets

from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))

def remove_stopwords(text):
    filtered_words = [
        word for word in text.split()
        if word.lower() not in stop_words
    ]
    return ' '.join(filtered_words)
```

# 8. Handling Emojis😂

```python
# two ways of handling emojis in text:
# 1. Remove emojis using regex
# 2. replace emojis with its meaning using emoji library
```

```python
def remove_emoji(text):
    emoji_pattern = re.compile("["
                           u"\U0001F600-\U0001F64F"  # emoticons
                           u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                           u"\U0001F680-\U0001F6FF"  # transport & map symbols
                           u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                           u"\U00002702-\U000027B0"
                           u"\U000024C2-\U0001F251"
                           "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', text)
```

```python
remove_emoji("Hello 😊! This is a test. 🚀")
```

# 10. Stemming

```python
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
def stem_words(text):
  return " ".join([ps.stem(word) for word in text.split()])
```

```python
sample = "walk walks walking walked"
```

```python
stem_words(sample)
```

```python
text = "probably my alltime favorite movie is story of selflessness sacrifice."
```

```python
print(text)
print(stem_words(text))
```

# 11. Lemmatization

```python
from nltk.stem import WordNetLemmatizer
wordnet_lemmatizer = WordNetLemmatizer()
```

```python
sentence = "He was running and eating at same time. He had bad habit of swimming after playing long hours in the Sun."
```

```python
tokens = word_tokenize(sentence)
```

```python
tokens
```

```python
import string
punctuations = string.punctuation
```

```python
punctuations
```

```python
for token in tokens:
  if token in punctuations:
    tokens.remove(token)
```

```python
# import nltk

# nltk.download('wordnet')
```

```python
print("{0:20}{1:20}".format("Word", "Lemma"))
for token in tokens:
  print("{0:20}{1:20}".format(token, wordnet_lemmatizer.lemmatize(token, pos = "v")))
```
