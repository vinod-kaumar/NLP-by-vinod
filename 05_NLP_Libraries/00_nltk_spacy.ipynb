## 1. NLTK (Natural Language Toolkit)
- NLTK is one of the oldest NLP libraries in Python.
- It was designed mainly for:
    - Teaching NLP
    - Research
    - Linguistics

```python
# 1. Tokenization
import nltk
# nltk.download('punkt_tab')
from nltk.tokenize import word_tokenize

text = "I love NLP"
print(word_tokenize(text))
```

```python
# 2. Sentence Tokenization
from nltk.tokenize import sent_tokenize

sent_tokenize(text)
```

```python
# 3. Stopwords
# nltk.download('stopwords')
from nltk.corpus import stopwords

stop_words = stopwords.words("english")
print(stop_words[:5])
```

```python
# 4. POS Tagging

# nltk.download('averaged_perceptron_tagger_eng') # new model
# nltk.download('averaged_perceptron_tagger') # old model

from nltk import pos_tag

pos_tag(["I", "love", "NLP"])
```

```python
# 5. Lemmatization

# nltk.download('wordnet')
# or
# nltk.download('omw-1.4') --> for multilingual WordNet support

from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()
lemmatizer.lemmatize("running", pos="v")
```

```python
# 6. NER

# nltk.download('maxent_ne_chunker_tab')
# nltk.download('words')

import nltk

# Step 1: Tokenize
tokens = nltk.word_tokenize(text)

# Step 2: POS Tagging
pos_tags = nltk.pos_tag(tokens)

# Step 3: Named Entity Recognition
entities = nltk.ne_chunk(pos_tags)

print(entities)
```

##### Advantages of NLTK
- Huge collection of NLP algorithms
- Excellent for learning
- Great documentation
- Many corpora included
#### Disadvantages of NLTK
- Slow
- Not optimized for production
- Requires combining many modules manually
- Older architecture

## 2. spaCy
- spaCy was built much later with one goal:
    - Fast industrial-strength NLP.
- When companies build NLP systems, spaCy is often preferred over NLTK.

```python
import spacy

nlp = spacy.load("en_core_web_sm")

doc = nlp("I love NLP.")
```

```python
# 1. tokenization
for token in doc:
    print(token.text)
```

```python
# 2. Lemmatization
for token in doc:
    print(token.text, token.lemma_)
```

```python
# 3. POS Tagging
for token in doc:
    print(token.text, token.pos_)
```

```python
# 4. NER
doc = nlp("Elon Musk founded SpaceX.")
for ent in doc.ents:
    print(ent.text, ent.label_)
```

```python
# 5. Dependency parsing -> shows grammatical relationships
for token in doc:
    print(token.text, token.dep_)
```

#### Advantages of spaCy
- Very fast
- Modern architecture
- Production-ready
- Excellent NER
- Efficient memory usage
- Easy API
#### Disadvantages of spaCy
- Fewer educational resources than NLTK
- Some advanced research tools absent
- Larger models consume more space

```python
# Spacy workflow

# Text
#   ↓
# Tokenizer
#   ↓
# POS Tagger
#   ↓
# Lemmatizer
#   ↓
# Dependency Parser
#   ↓
# NER
#   ↓
# Doc Object
```

##### Which should we learn?
1. Learn NLP concepts first:
    1. Tokenization
    2. Stemming
    3. Lemmatization
    4. POS Tagging
    5. N-grams
    6. TF-IDF
    - NLTK is great for this.
2. Learn practical NLP:
    1. NER
    2. Dependency Parsing
    3. Information Extraction
    - spaCy is better.
3. For modern NLP (what you're already studying):
    1. Word2Vec
    2. GloVe
    3. FastText
    4. BERT
    5. SBERT
    6. Transformers
    - Use Hugging Face Transformers more than either NLTK or spaCy.
