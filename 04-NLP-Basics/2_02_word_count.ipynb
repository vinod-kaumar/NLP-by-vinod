## Python Built-in Function - Count Words

```python
text = "Natural Language Processing (NLP) is a fascinating domain!"
words = text.split()
print(words)
print("word count:", len(words))
```

```python
def word_count(text):
    words = text.split()
    print("word count:",len(words) )
word_count(text)
word_count("Hello!, NLP is a fascinating field.")
word_count("Hello ! , NLP is a fascinating field .") # see this
```

## Regex - Handle Punctuation and Count Words

```python
import re
pattern = re.compile(r'\w+')
def word_count_re(text):
    words = re.findall(pattern, text)
    print(words)
    print("Word Count:", len(words))
word_count_re(text)
word_count_re("Hello!, NLP is a fascinating field.")
word_count_re("Hello ! , NLP is a fascinating field .") # see this
```

**Here we can see that RegEx is performing good as compare to split() method**

```python
from nltk import word_tokenize, sent_tokenize
def word_count_nltk(text):
    tokens = word_tokenize(text)
    print(tokens)
    print("Word count:", len(tokens))
def sentence_count_nltk(text):
    sentences = sent_tokenize(text)
    print(sentences)
    print("Sentences count:", len(sentences))
word_count_nltk("Hello!, NLP is a fascinating field.")
word_count_nltk("Hello ! , NLP is a fascinating field .") # see this
sentence_count_nltk("NLP is a fascinating domain. RegEx performs better than inbuilt python. We remove punctuations first then we use this method. so use Nltk ")
```

# SpaCy

```python
import spacy
nlp = spacy.load("en_core_web_sm")
def word_count_spacy(text):
    doc = nlp(text)
    tokens = [token.text for token in doc if not token.is_punct]
    print(tokens)
    print("Word count:", len(tokens))
word_count_spacy("Hello!, NLP is a fascinating field.")
word_count_spacy("Hello ! , NLP is a fascinating field .") # see this
```

# Counter 


**How many times a word came**

```python
from collections import Counter
def freq_count(text):
    words = text.lower().split()
    return Counter(words)
freq_count("NLP !,  is a fascinating field, nlp is fun. NLP is future .")
```

## NLP Task 1 - Short vs Long sentences

```python
import spacy
nlp = spacy.load("en_core_web_sm")

from nltk import sent_tokenize
def categorize_sentences(text):
    categories = {"long":[], "short":[]}
    sentences = sent_tokenize(text)
    for sentence in sentences:
        doc = nlp(sentence)
        # tokens = [token.text for token in doc] contains punctuations
        tokens = [token.text.lower() for token in doc if not token.is_punct]

        if(len(tokens) < 5):
            categories['short'].append(sentence)
        else :
            categories['long'].append(sentence)
    return categories        
text = "Hello! NLP is fun. Let's learn NLP together. It helps in many fields."
categories = categorize_sentences(text)
print("short sentences:", categories['short'])
print("long sentences:", categories['long'])
```

## NLP Task 2 - Rule based Sentiment Analysis(Heuristic Approach)

```python
from nltk import word_tokenize

positive_words = {"happy", "good", "great", "excellent", "love", "amazing"}
negative_words = {"sad", "bad", "terrible", "hate", "awful", "worst"}

import time

def sentiment_analysis(text):
    start = time.time()
    
    tokens = word_tokenize(text)
    
    pos_words = sum(1 for token in tokens if token in positive_words)
    neg_words = sum(1 for token in tokens if token in negative_words)
   
    sentiment = "Neutral Sentiment"
    
    if(pos_words > neg_words):
        sentiment = "Positive sentiment"    
    elif(neg_words > pos_words):
        sentiment = "Negative sentiment"
    
    end = time.time()
    exec_time = end - start
    return sentiment, exec_time
```

```python

text = """
    I'm happy you're the PM, and thank you for bringing me home from Ukraine.
    I have never been so proud to be an Indian citizen. Thank you, Mr. PM.
"""
sentiment, exec_time = sentiment_analysis(text)
print("Sentiment:", sentiment)
print("execution time:", exec_time)
```
